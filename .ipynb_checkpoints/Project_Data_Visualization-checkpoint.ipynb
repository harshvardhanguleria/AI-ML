```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\harsh\\Desktop\\covid_19_india.csv")
df.head(10)
```

```python
l = len(df.columns)
x = list()
for i in df.columns:
    x.append(i)
print(x)
print(type(x))
def pp(a, b, c):
    sns.pairplot(df, x_vars=[a,b,c],
                y_vars='Confirmed',
                height=5, aspect=1, kind = 'scatter')
    plt.show()
# a, b, c, d, e, f, g, h = x
pp(x[0], x[1], x[2])
pp(x[3], x[4], x[5])
pp(x[6], x[7], x[8])
print(l)
```
